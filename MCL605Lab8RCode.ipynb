```r
#prior probability that user is at location i,j
p = matrix(nrow=4, ncol=4, c(0.05,0.05,0.05,0.05,0.1,0.1,0.05,0.05,0.05,0.05,0.1,0.1,0.05,0.05,0.05,0.05))
print(p)
```

```r
#conditional probab observing 2 bar signal given user at i,j
b = matrix(nrow=4, ncol=4, c(0.75,0.05,0.01,0.01,0.95,0.75,0.05,0.01,0.75,0.95,0.75,0.05,0.05,0.75,0.95,0.75))
print(b)
```

```r
p %*% b # martix product
```

```r
p * b # component wise multiplication
```

```r
sum(p*b)
```

```r
#sudated probability 
(p*b)/sum(p*b)
```

```r
#sum sudated probability 
sum((p*b)/sum(p*b))
```
